```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Caminho para o arquivo GeoJSON original
arquivo_geojson = './Bairros_casos_lepto_2018_2019.geojson'

# Ler o arquivo GeoJSON com coordenadas UTM
gdf = gpd.read_file(arquivo_geojson)

# Verificar o sistema de coordenadas original (CRS)
print("Sistema de Coordenadas original (CRS):", gdf.crs)

# Converter as coordenadas UTM para latitude e longitude (WGS84 - EPSG:4326)
gdf_latlon = gdf.to_crs(epsg=4326)

# Verificar o novo sistema de coordenadas
print("Novo Sistema de Coordenadas (CRS):", gdf_latlon.crs)

# Caminho para o novo arquivo GeoJSON com coordenadas convertidas
novo_arquivo_geojson = './leptolat.geojson'

# Salvar o novo arquivo GeoJSON com coordenadas latitude/longitude
gdf_latlon.to_file(novo_arquivo_geojson, driver='GeoJSON')

# Plotar os polígonos com as novas coordenadas
gdf_latlon.plot()

# Ajustar o título e rótulos
plt.title('Polígonos com Coordenadas em Latitude/Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Mostrar o mapa
plt.show()

print(f"Novo arquivo GeoJSON salvo como: {novo_arquivo_geojson}")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Caminho para o arquivo GeoJSON original
arquivo_geojson = './Bairros_casos_lepto_2018_2019.geojson'

# Ler o arquivo GeoJSON com coordenadas UTM
gdf = gpd.read_file(arquivo_geojson)

# Verificar o sistema de coordenadas original (CRS)
print("Sistema de Coordenadas original (CRS):", gdf.crs)

# Converter as coordenadas UTM para latitude e longitude (WGS84 - EPSG:4326)
gdf_latlon = gdf.to_crs(epsg=4326)

# Filtrar o bairro chamado 'Centro'
bairro_centro = gdf_latlon[gdf_latlon['NomeBairro'] == 'Centro']

# Plotar todos os bairros em cinza
ax = gdf_latlon.plot(color='lightgray', edgecolor='black', figsize=(10, 10))

# Destacar o bairro 'Centro' em vermelho
bairro_centro.plot(ax=ax, color='red', edgecolor='black')

# Ajustar o título e rótulos
plt.title('Mapa dos Bairros com destaque para o Centro em vermelho')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Mostrar o mapa
plt.show()
```
